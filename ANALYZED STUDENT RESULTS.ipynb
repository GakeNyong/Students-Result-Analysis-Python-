```python
import pandas as pd
df = pd.read_excel("Results.xlsx")
df
del df["Total(100)"]
del df["Grade"]
del df["Comment"]
df
```

```python
df["Total(100)"] = df["Exam(65)"] + df["Midterm(20)"] + df["C.A(15)"]
df
#80-100, A
#70-79, B
#60-69, C
#50-59, D
#40-49, E
#30-39, F

df["Grade"] = df["Total(100)"].apply(lambda x: 'A' if x>=80 else 'B' if x>=70 else 'C' if x>=60 else 'D' if x>=50 else 'E' if x>=40 else 'F')
df

df["Comment"] = df["Total(100)"].apply(lambda x: 'Excellent' if x>=80
                                       else 'Very Good' if x>=70 
                                       else 'Good' if x>=60 
                                       else 'Fair' if x>=50 
                                       else 'Pass' if x>=40 
                                       else 'Fail')
df

decimals = 2
df["Exam(65)"] = df["Exam(65)"].apply(lambda x: round (x, decimals))
df

df=df.round(decimals=1)
df
```
